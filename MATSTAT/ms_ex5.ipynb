```r
set.seed(123)
```

```r
cer = read.csv("cereal.csv", sep = ";", header = TRUE, row.names = 1)
```

```r
cervals = cer[5:(ncol(cer)-5)][2:nrow(cer),]
cervals = scale(cervals)
```

```r
head(cervals)
```

Nejprve spočítám Euklidovské vzdálenosti pro nutriční údaje cereálií.

```r
?dist
```

```r
eucdist = dist(cervals,method = "euclidian")
```

Výstup funkce `dist` není v notebooku úplně přehledný, přepíšu ho tedy na matici.

```r
eucmat = as.matrix(eucdist)
eucmat
```

Nyní použiji na matici vzdáleností metodu MDS.

```r
?cmdscale
```

```r
mds = cmdscale(eucmat, k = 6)
```

```r
head(mds)
```

MDS nyní porovnám s metodou PCA.

```r
pca = prcomp(cervals, scale. = TRUE)
```

```r
?prcomp
```

```r
head(pca$x)
```

```r
corel = c()
for (i in 1:ncol(mds)){
    
    corel[i] = cor(mds[,i],pca$x[,i])
}
corel
```

Je patrné, že metodou MDS můžeme dojít ke stejným výsledkům jako přes PCA.

```r
nmmds = isoMDS(eucmat, k = 6)
```

```r
head(nmmds$points)
```

```r
corel2 = c()
for (i in 1:ncol(nmmds$points)){
    
    corel2[i] = cor(nmmds$points[,i],pca$x[,i])
}
corel2
```

Vidíme, že výsledek je opět totožný.
